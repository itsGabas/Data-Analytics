```python
def adicionar_item(itens, nome_item):
    item = {
        "nome": nome_item,
        "comprado": False
    }

    itens.append(item)
    print(f"O item {nome_item} foi adicionado com sucesso!")
    return

def ver_itens(itens):
    print("\nLista de Compras:")
    for indice, item in enumerate(itens, start=1):
        status = "✓" if item["comprado"] else " "
        nome_item = item["nome"]
        print(f"{indice}. [{status}] {nome_item}")
    return

def item_comprado(itens, indice_item):
    indice_item_ajustado = int(indice_item) - 1
    itens[indice_item_ajustado]["comprado"] = True
    print(f"Item {indice_item} foi comprado.")
    return

def remover_itens(itens):
    for item in itens:
      itens.remove(item)
    print("Item removido com sucesso!")
    return

itens = []

while True:
    print("\nMenu:")
    print("1. Adicionar Item")
    print("2. Listar Itens")
    print("3. Marcar Itens como comprados")
    print("4. Remover Itens")
    print("5. Sair")

    escolha = input("Digite a sua escolha: ")

    if escolha == '1':
        nome_item = input("Digite o nome da item que deseja comprar: ")
        adicionar_item(itens, nome_item)

    elif escolha == '2':
        ver_itens(itens)

    elif escolha == '3':
        ver_itens(itens)

        indice_item = input("Digite o número do item que deseja marcar como comprado: ")
        item_comprado(itens, indice_item)

    elif escolha == '4':
        remover_itens(itens)
        ver_itens(itens)

    elif escolha == '5':
        break

print("Encerrando sua Lista de Compras! Até mais.")
```
